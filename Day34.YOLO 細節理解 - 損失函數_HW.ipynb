## Day34.YOLO 細節理解 - 損失函數
今天的課程，我們講述了
* 損失函數是描述模型預測出來的結果和實際的差異的依據    
* YOLO 損失函數的設計包含物件位置的定位與物件類別辨識    
* YOLO損失函數透過超參數設定模型有不同的辨識能力    



### 作業
仔細觀察，bbox 寬高計算損失方式和bbox中心計算損失方式有哪邊不一樣嗎? 為什麼要有不同的設計?
![title](loss function.png)



你的答案:

希望指標能指出，大盒子中的小偏差比小盒子中的小偏差要小。 為了部分解決此問題，預測邊界框的寬度和高度的平方根，而不是直接預測寬度和高度。 換句話說，假設有兩個 ground truth，寬度和高度分別為 100 x 100 和 10 x 10，那麼相同的 5 個像素偏差會產生不同的影響。 因此作者使用平方根對此進行了補救。
